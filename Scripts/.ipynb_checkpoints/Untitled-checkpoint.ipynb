```python
import rasterio
import mplleaflet
import numpy as np
import matplotlib.pyplot as plt
from rasterio.plot import show
from pysheds.grid import Grid
```

```python
rstPath = '../Rst/20190109125130_1063922483.tif'
imgObj = rasterio.open(rstPath)
show(imgObj)
```

```python
#Open a digital elevation model 
grid = Grid.from_raster(rstPath, data_name='dem')
```

```python
imgObj.crs
```

```python
#Minnor slicing on borders to enhance colobars
elevDem=grid.dem[:-1,:-1]
```

```python
show(elevDem, cmap='Blues')
```

```python
# Detect depressions

# Detect depressions
depressions = grid.detect_depressions('dem')

# Plot depressions
plt.imshow(depressions, cmap='jet')
```

```python
# Fill depressions
grid.fill_depressions(data='dem', out_name='flooded_dem')

# Test result
depressions = grid.detect_depressions('flooded_dem')
plt.imshow(depressions, cmap='jet')
```

```python
# Detect flats
flats = grid.detect_flats('flooded_dem')

# Plot flats
plt.imshow(flats)
```

```python
grid.resolve_flats(data='flooded_dem', out_name='inflated_dem')
show(grid.inflated_dem[:-1,:-1])
```

```python
# Create a flow direction grid
#N    NE    E    SE    S    SW    W    NW
dirmap = (64,  128,  1,   2,    4,   8,    16,  32)

grid.flowdir(data='inflated_dem', out_name='dir', dirmap=dirmap)

show(grid.dir)
```

```python
# Specify discharge point
x, y = -107.91663,27.83479

# Delineate the catchment
grid.catchment(data='dir', x=x, y=y, dirmap=dirmap, out_name='catch',
               recursionlimit=15000, xytype='label', nodata_out=0)

# Clip the bounding box to the catchment
grid.clip_to('catch')

```

```python
# Get a view of the catchment
demView = grid.view('dem', nodata=np.nan)
show(demView)
grid.to_raster(demView, '../Output/clippedElevations_WGS84.tif')
```

```python
# Define the stream network

grid.accumulation(data='catch', dirmap=dirmap, 
                  pad_inplace=False, out_name='acc')

accView = grid.view('acc', nodata=np.nan)
show(accView)
```

```python
streams = grid.extract_river_network('catch', 'acc', threshold=200, dirmap=dirmap)

#streams[0]["geometry"]["coordinates"]
```

```python
#for a in (streams["features"]):
#    print(a)
```

```python
def saveDict(dic,file):
    f = open(file,'w')
    f.write(str(dic))
    f.close()
    
#save geojson as separate file
saveDict(streams,'../Output/streams_WGS84.geojson')
```

```python
from shapely.geometry import Polygon, mapping
shapes = grid.polygonize()
polyList = []
for shape in shapes:
    #print(shape[0]['coordinates'][0])
    #print('--------')
    cellPoly = Polygon(shape[0]['coordinates'][0])
    polyList.append(cellPoly)
```

```python
polyList[1]
```

```python
import fiona
outSchema = {
    'geometry':'Polygon',
    'properties':''
}

outFile = fiona.open('../Output/basinPolygons.geojson',mode = 'w',
                     driver = 'GeoJSON', schema=outSchema,
                    crs=imgObj.crs)
for index, poly in enumerate(polyList):
    outFile.write({
        'geometry':mapping(poly),
    })
outFile.close()
```
